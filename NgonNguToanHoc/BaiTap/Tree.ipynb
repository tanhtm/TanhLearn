```python
from queue import Queue, LifoQueue
from copy import deepcopy

class Node(object):
    
    def __init__ (self, value):
        self.value= value
        self.chil= []
    
    def add(self, node):
        self.chil.append(node)
    
    def bfs(self, returnNode= False):
        queue= Queue()
        queue.put(self)
        a= []
        while not queue.empty():
            node= queue.get()
            if returnNode:
                a.append(node)
            else:
                a.append((node.value, len(node.chil)))
            for c in node.chil:
                queue.put(c)
                
        return a
    
    def dfs(self, returnNode= False):
        stack= LifoQueue()
        stack.put(self)
        a= []
        while not stack.empty():
            node= stack.get()
            if returnNode:
                a.append(node)
            else:
                a.append((node.value, len(node.chil)))
            nodes= node.chil[::-1]
            for c in nodes :
                stack.put(c) 
        return a
        
    def show (self, bfs= True):
        a= self.bfs()
        j= 0
        for i in range(len(a)):
            val, l = a[i]
            c= [(k, a[k][0]) for k in range(j+1,j+l+1)]
            print('%d %s: %s' % (i, val, str(c)))
            j= j+l
                
    def copy(self):
        return deepcopy(self)
    
    def __getitem__(self, index):
        return self.chil[index]
    
class Tree:
        
    def __init__ (self, root):
        # root: l√† node g·ªëc c·ªßa c√¢y ƒë√£ t·ªìn t·∫°i
        self.root= root
        self.update()
    
    def update (self):
        # Duy·ªát danh s√°ch Node theo DFS
        self.nodes= self.root.dfs(returnNode= True)
        
    def show(self):
        self.root.show()
    
    def copy(self):
        return deepcopy(self)
    
    def __getitem__(self, index):
        return self.nodes[index]
```

```python
class Grammar:
    Name= 'G'
    Sigma= [] #T·∫≠p k√≠ hi·ªáu c∆° b·∫£n
    V= [] # T·∫≠p k√≠ hi·ªáu b·ªï sung
    sigma= 'ùúé' # K√Ω hi·ªáu xu·∫•t ph√°t c·ªßa vƒÉn ph·∫°m
    P= [] #rules C√°c quy t·∫Øc thay th·∫ø
    
    def __init__ (self, Sigma= [], V= [], sigma= 'ùúé', P= []):
        self.Sigma= Sigma
        self.V= V
        self.sigma= sigma
        self.P= P
    
    def getRules(self, w):
        return [r for r in self.P if r[0] == w]
    
    def display(self):
        print('%s= (Œ£,ùëâ,ùúé,ùëÉ)' % self.Name)
        print('Œ£= ' + str(self.Sigma))
        print('ùëâ= ' + str(self.V))
        print('ùúé= ' + str(self.sigma))
        tmp= [ '%s ‚Üí %s' % (str(rule[0]), str(rule[1])) for rule in self.P]
        print('ùëÉ= ' + '{' + ', '.join(tmp) + '}')
```

```python
G= Grammar()
G.Sigma= {'a', 'b'}
G.sigma= 'ùúé'
G.V= [G.sigma, 'A', 'B']
G.P= [(G.sigma, 'AB'), 
      ('A', 'aAb'), 
      ('A', 'b'), 
      ('B', 'bBa'), 
      ('B', None)]
G.display()
```

```python
def topDownParsing(G, w):
    
    def backtracking (tree, idNode, idW):
        print('--------------------------------')
        tree.show()
        # tree: C√¢y ·ªü tr·∫°ng th√°i hi·ªán t·∫°i
        # idNode: V·ªã tr√≠ node ƒëang x√©t tr√™n c√¢y, node n√†y l√† l√°
        # idW: V·ªã tr√≠ k√Ω t·ª± ƒëang x√©t trong w
        try:
            node= tree[idNode]
            T= node.value
        except:
            if idW >= len(w):
                return tree
            else:
                return None
        try:
            v= w[idW]
        except:
            return None
        print('T: %s, v: %s' % (str(T), v))
        if T in G.V:
            # T l√† k√≠ t·ª± b·ªï sung
            rules= G.getRules(T)
            for r in rules:
                treeCopy= tree.copy()
                A, B= r
                nodeCopy= treeCopy[idNode]
                if B is None: # K√≠ t·ª± r·ªóng
                    nodeCopy.add(Node(None))
                    treeCopy.update()
                else:
                    for c in B:
                        nodeCopy.add(Node(c))
                    treeCopy.update()
                resultTree= backtracking(treeCopy, idNode + 1, idW)
                if not resultTree is None:
                    return resultTree
                    
        elif T is None:
            tree.show()
            resultTree= backtracking(tree, idNode + 1, idW)
            if not resultTree is None:
                return resultTree
        else:
            # T l√† k√≠ t·ª± c∆° b·∫£n
            if T == v:
                resultTree= backtracking(tree, idNode + 1, idW + 1)
                if not resultTree is None:
                    return resultTree
            else:
                # Quay lui
                return None    
        
    
    sigma= G.sigma
    for  r in G.getRules(sigma):
        node= Node(r[0])
        if r[1] is None: # sigma -> None th√¨ v√¥ nghƒ©a
            continue
        for c in r[1]:
            node.add(Node(c))
        tree= Tree(node)
        resultTree= backtracking(tree, idNode= 1, idW= 0)
        if not resultTree is None:
            return resultTree
    
    return None
```

```python
G= Grammar()
G.Sigma= {'a', 'b'}
G.sigma= 'ùúé'
G.V= [G.sigma, 'A', 'B']
G.P= [(G.sigma, 'AB'), 
      ('A', 'aAb'), 
      ('A', 'b'), 
      ('B', 'bBa'), 
      ('B', None)]
G.display()
```

```python
r= topDownParsing(G, 'abbba')
```

```python
if r is None:
    print('No solution !')
else:
    r.show()
```
