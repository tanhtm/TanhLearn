```python
from queue import Queue, LifoQueue
from copy import deepcopy

class Node(object):
    
    def __init__ (self, value):
        self.value= value
        self.chil= []
    
    def add(self, node):
        self.chil.append(node)
    
    def bfs(self, returnNode= False):
        queue= Queue()
        queue.put(self)
        a= []
        while not queue.empty():
            node= queue.get()
            if returnNode:
                a.append(node)
            else:
                a.append((node.value, len(node.chil)))
            for c in node.chil:
                queue.put(c)
                
        return a
    
    def dfs(self, returnNode= False):
        stack= LifoQueue()
        stack.put(self)
        a= []
        while not stack.empty():
            node= stack.get()
            if returnNode:
                a.append(node)
            else:
                a.append((node.value, len(node.chil)))
            nodes= node.chil[::-1]
            for c in nodes :
                stack.put(c) 
        return a
        
    def show (self, bfs= True):
        a= self.bfs()
        j= 0
        for i in range(len(a)):
            val, l = a[i]
            c= [(k, a[k][0]) for k in range(j+1,j+l+1)]
            print('%d %s: %s' % (i, val, str(c)))
            j= j+l
                
    def copy(self):
        return deepcopy(self)
    
    def __getitem__(self, index):
        return self.chil[index]
    
class Tree:
        
    def __init__ (self, root):
        # root: lÃ  node gá»‘c cá»§a cÃ¢y Ä‘Ã£ tá»“n táº¡i
        self.root= root
        self.update()
    
    def update (self):
        # Duyá»‡t danh sÃ¡ch Node theo DFS
        self.nodes= self.root.dfs(returnNode= True)
        
    def show(self):
        self.root.show()
    
    def copy(self):
        return deepcopy(self)
    
    def __getitem__(self, index):
        return self.nodes[index]
```

```python
class Grammar:
    Name= 'G'
    Sigma= [] #Táº­p kÃ­ hiá»‡u cÆ¡ báº£n
    V= [] # Táº­p kÃ­ hiá»‡u bá»• sung
    sigma= 'ğœ' # KÃ½ hiá»‡u xuáº¥t phÃ¡t cá»§a vÄƒn pháº¡m
    P= [] #rules CÃ¡c quy táº¯c thay tháº¿
    
    def __init__ (self, Sigma= [], V= [], sigma= 'ğœ', P= []):
        self.Sigma= Sigma
        self.V= V
        self.sigma= sigma
        self.P= P
    
    def getRules(self, w):
        return [r for r in self.P if r[0] == w]
    
    def display(self):
        print('%s= (Î£,ğ‘‰,ğœ,ğ‘ƒ)' % self.Name)
        print('Î£= ' + str(self.Sigma))
        print('ğ‘‰= ' + str(self.V))
        print('ğœ= ' + str(self.sigma))
        tmp= [ '%s â†’ %s' % (str(rule[0]), str(rule[1])) for rule in self.P]
        print('ğ‘ƒ= ' + '{' + ', '.join(tmp) + '}')
```

```python
def topDownParsing(G, w):
    
    def backtracking (tree, idNode, idW):
        print('--------------------------------')
        tree.show()
        # tree: CÃ¢y á»Ÿ tráº¡ng thÃ¡i hiá»‡n táº¡i
        # idNode: Vá»‹ trÃ­ node Ä‘ang xÃ©t trÃªn cÃ¢y, node nÃ y lÃ  lÃ¡
        # idW: Vá»‹ trÃ­ kÃ½ tá»± Ä‘ang xÃ©t trong w
        try:
            node= tree[idNode]
            T= node.value
        except:
            if idW >= len(w):
                return tree
            else:
                return None
        try:
            v= w[idW]
        except:
            return None
        print('T: %s, v: %s' % (str(T), v))
        if T in G.V:
            # T lÃ  kÃ­ tá»± bá»• sung
            rules= G.getRules(T)
            for r in rules:
                treeCopy= tree.copy()
                A, B= r
                nodeCopy= treeCopy[idNode]
                if B is None: # KÃ­ tá»± rá»—ng
                    nodeCopy.add(Node(None))
                    treeCopy.update()
                else:
                    for c in B:
                        nodeCopy.add(Node(c))
                    treeCopy.update()
                resultTree= backtracking(treeCopy, idNode + 1, idW)
                if not resultTree is None:
                    return resultTree
                    
        elif T is None:
            tree.show()
            resultTree= backtracking(tree, idNode + 1, idW)
            if not resultTree is None:
                return resultTree
        else:
            # T lÃ  kÃ­ tá»± cÆ¡ báº£n
            if T == v:
                resultTree= backtracking(tree, idNode + 1, idW + 1)
                if not resultTree is None:
                    return resultTree
            else:
                # Quay lui
                return None    
        
    
    sigma= G.sigma
    for  r in G.getRules(sigma):
        node= Node(r[0])
        if r[1] is None: # sigma -> None thÃ¬ vÃ´ nghÄ©a
            continue
        for c in r[1]:
            node.add(Node(c))
        tree= Tree(node)
        resultTree= backtracking(tree, idNode= 1, idW= 0)
        if not resultTree is None:
            return resultTree
    
    return None
```

```python
G= Grammar()
G.Sigma= {'a', 'b'}
G.sigma= 'ğœ'
G.V= [G.sigma, 'A', 'B']
G.P= [(G.sigma, 'AB'), 
      ('A', 'aAb'), 
      ('A', 'b'), 
      ('B', 'bBa'), 
      ('B', None)]
G.display()
```

```python
r= topDownParsing(G, 'abbba')
```

```python
if r is None:
    print('No solution !')
else:
    r.show()
```
