```python
import matplotlib.pyplot as plt

from Crypto.Signature import PKCS1_v1_5
from Crypto.Hash import SHA
from Crypto.PublicKey import RSA
import Crypto
import binascii
```

```python
# Lấy ảnh và sửa

fig, ax = plt.subplots(2, 1,figsize=(20,20))
im= plt.imread('image.png')
ax[0].imshow(im)

im_= im
im_[100, 100, 0]= (im_[100, 100, 0] + 1)/2 # Thay đổi giá trị 1 pixel
plt.imsave('image_.png', im_ , format= 'PNG')
ax[1].imshow(im_)
```

```python
import hashlib
def hashFile(file):
    BLOCK_SIZE = 65536 # The size of each read from the file
    file_hash = hashlib.sha256()
    with open(file, 'rb') as f: #read bytes
        fb = f.read(BLOCK_SIZE)
        while len(fb) > 0:
            file_hash.update(fb)
            fb = f.read(BLOCK_SIZE)
    return file_hash
```

```python
# Băm ảnh gốc

file = "image.png"
file_hash= hashFile(file)
file_hash.hexdigest()
```

```python
# Băm ảnh sửa

_file = "image_.png"
_file_hash= hashFile(_file)
_file_hash.hexdigest()
```

```python
# Ký và xác minh chữ ký số
file = "image.png"

# Tạo khóa để ký
random = Crypto.Random.new().read
private_key = RSA.generate(1024, random)
signer = PKCS1_v1_5.new(private_key)

# Băm
h= hashFile(file).digest()
h = SHA.new(h)

#Ký
sign= signer.sign(h)
sign
```

```python
# Xác minh chữ ký
verifier = PKCS1_v1_5.new(private_key)
verifier.verify(h, sign)
```

```python
# Tạo khóa khác để kiểm tra
private_key2 = RSA.generate(1024, random)

verifier = PKCS1_v1_5.new(private_key2)
verifier.verify(h, sign)
```
